```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
#RANDOMWALK 1D
def Randomwalk1D(n):
    x = 0 
    y = 0 
    xf = [0]
    yf = [0]

    for i in range(1,n+1):
        step = np.random.uniform(0,1)

        if step < 0.5:
            x += 1
            y += 1
        else:
            x += 1 
            y += -1
        xf.append(x)
        yf.append(y)

    return [xf,yf]


n = 1000
Randwalk = Randomwalk1D(n)
Randwalk = np.array(Randwalk)

msd = np.zeros(1001)
pos = 0.
for i,j in enumerate(Randwalk[1][:]):
    pos = pos + np.sqrt(j*j)
    msd[i] = pos 


plt.figure(figsize=(10,8))
plt.plot(Randwalk[0],Randwalk[1],'r-', color='#e000b7', markersize=8)
plt.title("Caminhante Aleat√≥rio 1D - p=0.5")
plt.xlabel("Passos")
#plt.savefig('rwp05.png', format='png')
plt.show()

plt.figure(figsize=(10,8))
plt.plot(Randwalk[0][:],msd,'r-',color='#0524ed', markersize=8)
plt.title("Deslocamento Quadr√°tico M√©dio (MSD)")
plt.xlabel("Tempo")
plt.ylabel("MSD")
#plt.savefig('msdp05.png', format='png')
plt.show()
```

```python
#RANDOMWALK 2D
def Randomwalk2D(n):
    x = np.zeros(n)
    y = np.zeros(n)
    directions = ["UP", "DOWN", "LEFT", "RIGHT"]
    for i in range(1, n):
        step = random.choice(directions)

        if step == "RIGHT":
            x[i] = x[i - 1] + 1
            y[i] = y[i - 1]
        elif step == "LEFT":
            x[i] = x[i - 1] - 1
            y[i] = y[i - 1]
        elif step == "UP":
            x[i] = x[i - 1]
            y[i] = y[i - 1] + 1
        elif step == "DOWN":
            x[i] = x[i - 1]
            y[i] = y[i - 1] - 1
    
    return x, y

n = 5000
Randwalk2d = Randomwalk2D(n)
Randwalk2d = np.array(Randwalk2d)
rw1 = Randomwalk2D(5000)
rw2 = Randomwalk2D(5000)
rw3 = Randomwalk2D(5000)
rw4 = Randomwalk2D(5000)
rw5 = Randomwalk2D(5000)
rw6 = Randomwalk2D(5000)

msd = np.zeros(5000)
pos = 0.
for i,j in enumerate(Randwalk2d[1][:]):
    pos = pos + np.sqrt(j*j)
    msd[i] = pos 


plt.figure(figsize=(10,8))
plt.plot(Randwalk2d[0],Randwalk2d[1], color='#6c02bd', markersize=8)
plt.plot(rw1[0], rw1[1])
plt.plot(rw2[0], rw2[1])
plt.plot(rw3[0], rw3[1])
plt.plot(rw4[0], rw3[1])
plt.plot(rw5[0], rw3[1])
plt.plot(rw6[0], rw3[1])
plt.title("Caminhante Aleat√≥rio 2D")
#plt.savefig('rw2d.png', format='png')
plt.show()

plt.figure(figsize=(10,8))
plt.plot(Randwalk2d[0][:],msd,'r-',color='#0593ff', markersize=8)
plt.title("Deslocamento Quadr√°tico M√©dio (MSD)")
plt.xlabel("Tempo")
plt.ylabel("MSD")
#plt.savefig('msd2d.png', format='png')
plt.show()
```
